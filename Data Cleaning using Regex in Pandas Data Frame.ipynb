# Data Cleaning using Regex in Pandas DataFrame

```python
import re
import pandas as pd
```

```python
logs = pd.read_csv('datasets/log.csv')
logs.head()
```

```python
# Search for all video columns that ends with html and replace it with webpage.

# \w- accept all values whether its numerical or alphabet or integer
# $ - to check where the value ends with something
```

```python
logs.replace(to_replace = '[.]html$', value = '',regex = True)
```

```python
df = pd.read_csv("datasets/presidents.csv")
```

```python
df
```

```python
r = 'George W Bush'

re.findall('\w+',r)
```

```python
re.findall('[\w-]+$',r)
```

```python
re.findall('(?:[\w]+\s)(\w+[.]?)(?:\s\w+)',r) #?: match and replace , ? expect 0 or 1 characeter
```

```python
df['First Name'] = df['President'].str.extract('(\w+)')
df['Middle Name'] = df['President'].str.extract('(?:[\w]+\s)(\w+[.]?)(?:\s\w+)')
df['Last Name'] = df['President'].str.extract('([\w-]+$)')
```

```python
df
```

```python
r = 'Oct 30, 1735[a]'

re.findall('\w{1,3}\s\d{1,2}[,]\s\d{1,4}',r)
```

```python
df['Born'] = df['Born'].str.extract('(\w{1,3}\s\d{1,2}[,]\s\d{1,4})')
```

```python
df['Year of Presidency'] = df['Age atstart of presidency'].str.extract('(\d{1,2})(?:\s)(?:\w+)(?:,)')
df['Days of Presidency'] = df['Age atstart of presidency'].str.extract('(?:\d{1,2})(?:\s)(?:\w+)(?:,)(?:\s)(\d{1,10})')
df['Date of Presidency'] = df['Age atstart of presidency'].str.extract('(?:\d{1,2})(?:\s)(?:\w+)(?:,)(?:\s)(?:\d{1,10})(?:\s)(?:\w{4})(\w.*)')
```
